```python
from ocr_complete import wrapper
import os
from tqdm import tqdm
from word_processing import editDistance
```

```python
folder_path = 'Patient_Records'
med_dict = {}
for root, dirs, files in os.walk(folder_path):
    for file in tqdm(files):
        file_path = os.path.join(root, file)
        if file.endswith(('.jpg', '.jpeg', '.png')):
            result = wrapper(file_path,min_len=3,closest_med=1)
            
            for words in result.values():
                med_words = words[1]
                for med_term in med_words:
                    if med_term in med_dict:
                        med_dict[med_term]+= 1
                        
                    else:
                        med_dict[med_term]=1
```

```python
med_dict
```

```python
import matplotlib.pyplot as plt



# Sorting words by frequency to get the top 10 most frequent words
sorted_words = sorted(med_dict.items(), key=lambda x: x[1], reverse=True)[:10]

# Extracting words and their frequencies
top_words, frequencies = zip(*sorted_words)

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.barh(top_words, frequencies, color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.title('Top 10 Most Frequent Words and Their Frequencies')
plt.gca().invert_yaxis()  # Invert y-axis to display highest frequency at the top
plt.tight_layout()
plt.show()
```
